```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime
from dash import Dash, html, dcc
```

```python
df = pd.read_csv('weather.csv')#, index_col = 'datetime', parse_dates = True)
df.drop(df.columns[[0]],axis = 1, inplace = True)
```

```python
fig = px.bar(df, x= 'datetime', y = 'precip', labels = {'precip' : 'Precipitation (inches)', 'datetime' : 'Date'})
fig.update_layout(
    title={
        'text': 'Amount of Rainfall Throughout 2022',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})


```

```python
import plotly.graph_objects as go


# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['datetime'], y=df['temp'],
                    mode='lines',
                    name='Average Temp'))
fig.add_trace(go.Scatter(x=df['datetime'], y=df['tempmax'],
                    mode='lines',
                    name='Max Temp'))
fig.add_trace(go.Scatter(x=df['datetime'], y=df['tempmin'],
                    mode='lines', name='Min Temp'))

fig.show()
```

```python
fig = go.Figure()
# Create and style traces
fig.add_trace(go.Scatter(x=df['datetime'], y=df['temp'], name='Average Temp',
                         line=dict(color='green', width=3)))
fig.add_trace(go.Scatter(x=df['datetime'], y=df['tempmin'], name = 'Min Temp',
                         line=dict(color='royalblue', width=2, dash = 'dot')))
fig.add_trace(go.Scatter(x=df['datetime'], y=df['tempmax'], name='Max Temp',
                         line=dict(color='firebrick', width=2,
                              dash='dot'))) # dash options include 'dash', 'dot', and 'dashdot'))


# Edit the layout
fig.update_layout(title='Average, High, and Low Temperatures in Provo',
                   xaxis_title='Month',
                   yaxis_title='Temperature (degrees F)')
```

```python
#update table 

from dash import Dash, dash_table
import pandas as pd

df = pd.read_csv('weather.csv')#, index_col = 'datetime', parse_dates = True)
df.drop(df.columns[[0]],axis = 1, inplace = True)

app = Dash(__name__)

app.layout = dash_table.DataTable(
    df.to_dict("records"),
    [{"name": i, "id": i} for i in df.columns],
    filter_action="native",
    filter_options={"placeholder_text": "Filter column..."},
    page_size=10,
)

if __name__ == "__main__":
    app.run_server(debug=True)
```
